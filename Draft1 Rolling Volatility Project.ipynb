```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf
from datetime import date
yf.pdr_override()
from jupyterthemes import get_themes
import jupyterthemes as jt
from jupyterthemes.stylefx import set_nb_theme
set_nb_theme('monokai')
import itertools

#GOAL- Use the rolling volatility, 1 year, 252 trading days, calculate rolling vol
#vol is sd of returns btw
#For a given stock index, calculate the rolling volatilities for all stocks in that index
#Take top ten highest and lowest 10- these are the ones you long and short respectively
#Do thing above for every 3 months
```

```python
pd.read_csv("SP500.csv")
data = pd.read_csv("SP500.csv")
symbols = np.array(data["Symbol"])
```

```python
def longTopShortBottom1stQ():
    rollingVolList = []
    TICKR = []
    for ticker in symbols: #Downloads data for each ticker and then appends each ticker to a list. Calculates rolling vol and then gets mean.
        dataValue = yf.download(ticker, start="2021-01-01", end="2021-04-01")
        TICKR.append(ticker)
        dataValue = dataValue["Close"]
        rollingVol = dataValue.pct_change().rolling(20).std()*(252**0.5)
        rollingVolList.append(rollingVol.mean())
    
    #Statements to print out stocks to long and short. I use the code below to sort the ticker list and rollingVolList in the same way
    rollingVolList, TICKR = zip(*sorted(zip(rollingVolList, TICKR)))
    print(f'You should long:{TICKR[0], TICKR[1], TICKR[2], TICKR[3], TICKR[4], TICKR[5], TICKR[6], TICKR[7], TICKR[8], TICKR[9]}')
    print(f'You should short:{TICKR[495], TICKR[496], TICKR[497], TICKR[498], TICKR[499], TICKR[500], TICKR[501], TICKR[502], TICKR[503], TICKR[504],}')

def longTopShortBottom2ndQ():
    rollingVolList = []
    TICKR = []
    for ticker in symbols:
        dataValue = yf.download(ticker, start="2021-04-01", end="2021-07-01")
        TICKR.append(ticker)
        dataValue = dataValue["Close"]
        rollingVol = dataValue.pct_change().rolling(20).std()*(252**0.5)
        rollingVolList.append(rollingVol.mean())
    
    rollingVolList, TICKR = zip(*sorted(zip(rollingVolList, TICKR)))
    print(f'You should long:{TICKR[0], TICKR[1], TICKR[2], TICKR[3], TICKR[4], TICKR[5], TICKR[6], TICKR[7], TICKR[8], TICKR[9]}')
    print(f'You should short:{TICKR[495], TICKR[496], TICKR[497], TICKR[498], TICKR[499], TICKR[500], TICKR[501], TICKR[502], TICKR[503], TICKR[504],}')  

def longTopShortBottom3rdQ():
    rollingVolList = []
    TICKR = []
    for ticker in symbols:
        dataValue = yf.download(ticker, start="2021-07-01", end="2021-10-01")
        TICKR.append(ticker)
        dataValue = dataValue["Close"]
        rollingVol = dataValue.pct_change().rolling(20).std()*(252**0.5)
        rollingVolList.append(rollingVol.mean())

    rollingVolList, TICKR = zip(*sorted(zip(rollingVolList, TICKR)))
    print(f'You should long:{TICKR[0], TICKR[1], TICKR[2], TICKR[3], TICKR[4], TICKR[5], TICKR[6], TICKR[7], TICKR[8], TICKR[9]}')
    print(f'You should short:{TICKR[495], TICKR[496], TICKR[497], TICKR[498], TICKR[499], TICKR[500], TICKR[501], TICKR[502], TICKR[503], TICKR[504],}')
    
def longTopShortBottom4thQ():
    rollingVolList = []
    TICKR = []
    for ticker in symbols: 
        dataValue = yf.download(ticker, start="2021-10-01", end="2022-01-01")
        TICKR.append(ticker)
        dataValue = dataValue["Close"]
        rollingVol = dataValue.pct_change().rolling(20).std()*(252**0.5)
        rollingVolList.append(rollingVol.mean())
    
    rollingVolList, TICKR = zip(*sorted(zip(rollingVolList, TICKR)))
    print(f'You should long:{TICKR[0], TICKR[1], TICKR[2], TICKR[3], TICKR[4], TICKR[5], TICKR[6], TICKR[7], TICKR[8], TICKR[9]}')
    print(f'You should short:{TICKR[495], TICKR[496], TICKR[497], TICKR[498], TICKR[499], TICKR[500], TICKR[501], TICKR[502], TICKR[503], TICKR[504],}')
```
